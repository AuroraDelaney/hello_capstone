```python
happiness = pd.read_csv('../data/world-happiness-report-2021.csv')
happiness = happiness.reset_index(drop=True)
happiness.head()
```

```python
depression = pd.read_csv('../data/share-with-depression.csv')
depression = depression.reset_index(drop=True)
depression.head(15)
```

```python
prevdep = pd.read_csv('../data/prevalence-of-depression-males-vs-females.csv')
prevdep = prevdep.reset_index(drop=True)
prevdep.head(15)
```

```python
prevdep.rename(columns={'Prevalence - Depressive disorders - Sex: Male - Age: Age-standardized (Percent)': 'prevalence of DD by male', 'Prevalence - Depressive disorders - Sex: Female - Age: Age-standardized (Percent)': 'prevalence of DD by female'}, inplace=True)
```

```python
feelings = pd.concat([happiness, depression, prevdep])
feelings.head(15)
```

```python
happiness = pd.read_csv('../data/world-happiness-report-2021.csv')
happiness = happiness.reset_index(drop=True)
happiness.head(15)
```

```python
depression = pd.read_csv('../data/share-with-depression.csv')
depression = depression.reset_index(drop=True)
depression.head(15)
```

```python
prevdep = pd.read_csv('../data/prevalence-of-depression-males-vs-females.csv')
prevdep = prevdep.reset_index(drop=True)
prevdep.head(15)
```

```python
feelings = pd.concat([happiness, depression])
feelings.head(15)
```

```python
feelings.rename(columns={'Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)': 'Prevalence- depressive orders - sexs and age precentage'}, inplace=True)
```

```python
feelings.groupby(['Country name', 'Healthy life expectancy'])['Prevalence- depressive orders - sexs and age precentage'].count().sort_values(ascending=False)
```

```python
feelings.groupby(['Country name', 'Healthy life expectancy'])['Prevalence- depressive orders - sexs and age precentage'].count().sort_values(ascending=False).unstack()
```
